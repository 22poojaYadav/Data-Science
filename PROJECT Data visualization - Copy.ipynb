```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('students result.csv')
print(df)
```

```python
plt.figure(figsize=(6,5))
plt.plot(df['Name'],df['Enghlish'])
plt.title('students marks')
plt.xlabel('students Name')
plt.ylabel('Enghlish')
plt.show()
```

```python
plt.figure(figsize=(6,5))
plt.bar(df['Name'],df['Math'])
plt.title('students Marks')
plt.xlabel('students Name')
plt.ylabel('Math')
plt.show()
```
